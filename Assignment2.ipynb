```python
import numpy as np
from scipy import stats
z = np.abs(stats.zscore(df['Math_Score']))
print(z)
```

```python
threshold = 0.18
sample_outliers = np.where(z <threshold)
sample_outliers
```

```python
sorted_score = sorted(df['Reading_Score'])
print(sorted_score)
```

```python
 q1 = np.percentile(sorted_score, 25)
 q3 = np.percentile(sorted_score, 75)
 print(q1,q3)
```

```python
 iqr = q3-q1
 lbound = q1-(1.5*iqr)
 ubound = q3+(1.5*iqr)
 print(lbound, ubound)
```

```python
 r_outliers = []
 for i in sorted_score:
     if (i<lbound or i>ubound):
         r_outliers.append(i)
 print(r_outliers)
```
